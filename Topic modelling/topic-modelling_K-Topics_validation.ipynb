```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame
df = pd.read_csv('tm_scores/coherence-scores-v2.csv', sep=';')

```

### Umass coherence score

```python
# Extract the number of topics and coherence scores
num_topics = df['Topics']
coherence_umass_scores = df['Coherence score u_mass']

# Plot the graph
plt.plot(num_topics, coherence_umass_scores, marker='o', label='Coherence score Umass')
plt.xlabel('Number of Topics')
plt.ylabel('Coherence Score')
plt.title('Umass Coherence Score vs. Number of Topics')
plt.grid(True)

# Set the x-axis limits
plt.xlim(left=0, right=100)

# Adjust the layout to prevent the title from being cut off
plt.tight_layout()

plt.grid(True)
plt.legend()

# Save the graph as a PNG image
plt.savefig('tm_scores/tm_coherence_Umass_graph.png')

# Show the graph
plt.show()
```

### CV coherence score

```python
# Extract the number of topics and coherence scores
num_topics = df['Topics']
coherence_cv_scores = df['Coherence score c_v']

# Plot the graph
plt.plot(num_topics, coherence_cv_scores, marker='o', color='orange', label='Coherence score c_v')
plt.xlabel('Number of Topics')
plt.ylabel('Coherence Score')
plt.title('CV Coherence Score vs. Number of Topics')
plt.grid(True)

# Set the x-axis limits
plt.xlim(left=0, right=100)

# Adjust the layout to prevent the title from being cut off
plt.tight_layout()

plt.grid(True)
plt.legend()

# Save the graph as a PNG image
plt.savefig('tm_scores/tm_coherence_cv_graph.png')

# Show the graph
plt.show()
```

### Perplexity score

```python
# Extract the number of topics and coherence scores
num_topics = df['Topics']
perplexity_scores = df['Perplexity score']

# Plot the graph
plt.plot(num_topics, perplexity_scores, color='red', marker='o', label='Perplexity score')
plt.xlabel('Number of Topics')
plt.ylabel('Perplexity Score')
plt.title('Perplexity Scores vs. Number of Topics')
plt.grid(True)

# Set the x-axis limits
plt.xlim(left=0, right=100)

# Adjust the layout to prevent the title from being cut off
plt.tight_layout()

# Reverse the y-axis direction
plt.gca().invert_yaxis()

plt.grid(True)
plt.legend()

# Save the graph as a PNG image
plt.savefig('tm_scores/perplexity-scores_graph.png')

# Show the graph
plt.show()
```

## Human intrepetation scores

```python
# Read the CSV file into a DataFrame
df_human = pd.read_csv('tm_scores/tm_human-intrepretation.csv', sep=';')
```

```python
df_human.head(5)
```

```python
# Extract the number of topics and coherence scores
num_topics = df_human['K topics']
solid_topics = df_human['Solid topics']
bad_topics = df_human['Bad topics']
undecided_topics = df_human['Undecided topics']

# Plot the graph
plt.plot(num_topics, solid_topics, color='green', marker='o', label='Coherent topics')
plt.plot(num_topics, bad_topics, color='red', marker='o', label='Incoherent topics')
plt.plot(num_topics, undecided_topics, color='orange', marker='o', label='Undecided topics')
plt.xlabel('K topics')
plt.ylabel('Number of Topics')
plt.title('Coherent, Incoherent and undecided topics according to human interpretation')
plt.grid(True)

# Set the x-axis limits
plt.xlim(left=0, right=50)
plt.ylim(bottom=0)
# Adjust the layout to prevent the title from being cut off
plt.tight_layout()

plt.grid(True)
plt.legend()

# Save the graph as a PNG image
plt.savefig('tm_scores/human-interpretation.png')

# Show the graph
plt.show()
```

## Independent coherence scores per topic

```python
df_topic = pd.read_csv('ds_test/topic_models_distribution/topic_coherence-scores.csv', low_memory=False)
```

```python
import matplotlib.patches as mpatches

# Extract the number of topics and coherence scores
num_topics = df_topic['topic_number']
topic_numbers = 30
coherence_scores = df_topic['coherence_score']

# Plot the graph
colors = []
for topic_number in num_topics:
    if topic_number in [1, 2, 4, 5, 6, 7, 9, 10, 14, 15, 17, 18, 19, 20, 22, 23, 27, 29]:
        colors.append('green')
    elif topic_number in [3, 12, 25, 26]:
        colors.append('orange')
    else:
        colors.append('red')

plt.figure(figsize=(10, 6))  # Adjust the figure size as needed
plt.bar(num_topics, coherence_scores, width=0.75, color=colors)
plt.xlabel('Number of Topics')
plt.ylabel('Coherence Score')
plt.title('CV Coherence Score per topic')
plt.xlim(min(num_topics)-0.5, max(num_topics)+0.5)  # Add a margin of 0.5 on both ends


# Set custom labels for each bar on the x-axis
plt.xticks(range(len(num_topics)), num_topics)

# Adjust the layout to prevent the title from being cut off
plt.tight_layout()


# Save the graph as a PNG image
plt.savefig('tm_scores/tm_topic-coherence_graph.png')

# Show the graph
plt.show()
```
