```python
import pandas as pd
import numpy as np
```

```python
# Read the CSV file into a DataFrame
topicnumber = 23
data = pd.read_csv(f'ds_topics/tm_topic_{topicnumber}_data.csv', low_memory=False, lineterminator='\n')

seed = 132432

# Filter comments with at least 5 words
data['word_count'] = data['body'].str.split().apply(lambda x: len(x))
data = data[data['word_count'] >= 10]

# Select random neutral comments
neutral_comments = data[data['sentiment_vader'] == 'neutral'].sample(n=20, random_state=seed)

# Select random positive comments
positive_comments = data[data['sentiment_vader'] == 'positive'].sample(n=20, random_state=seed)

# Select random negative comments
negative_comments = data[data['sentiment_vader'] == 'negative'].sample(n=20, random_state=seed)

# Concatenate the selected comments into a new dataframe
selected_comments = pd.concat([neutral_comments, positive_comments, negative_comments])

# Reset the index of the new dataframe
selected_comments = selected_comments.reset_index(drop=True)

selected_comments.to_csv(f'ds_stressfactor/tm_topic_{topicnumber}_subdataset.csv', index=False)

selected_comments.head(5)

```
