```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('ds_master/reddit_climatechange_dataset_labelled.csv', low_memory=False, lineterminator='\n')
```

```python
# Define the topic numbers to include
topic_numbers_to_include = [1, 4, 5, 6, 7, 9, 10, 14, 15, 18, 19, 20, 22, 23]

# Filter out the data points
climate_df = df[df['top_weighted_topic'].isin(topic_numbers_to_include)]

```

```python
climate_df.head(50)
```

```python
# Create an empty DataFrame to store the filtered results
# filtered_df = pd.DataFrame(columns=['body'])

# Filter the comments containing 'stress' or 'anxiety'
filtered_inducing_df = climate_df[climate_df['body'].str.contains('stress|anxiety', case=False, regex=True)]
filtered_reducing_df = climate_df[climate_df['body'].str.contains('relief|comfort', case=False, regex=True)]
```

```python
print(len(filtered_inducing_df), ' ', len(filtered_reducing_df))
```

```python
analyse_inducing_comments = filtered_inducing_df.sample(n=250)
analyse_reducing_comments = filtered_reducing_df.sample(n=250)
```

```python
analyse_inducing_comments.head(50)
```

```python
filtered_inducing_df.to_csv(f'ds_thematic-analysis/ds_stress-inducing-filtered_comments.csv', index=False)
analyse_inducing_comments.to_csv(f'ds_thematic-analysis/ds_stress-inducing-filtered_250comments_analysis.csv', index=False)
filtered_reducing_df.to_csv(f'ds_thematic-analysis/ds_stress-reducing-filtered_comments.csv', index=False)
analyse_reducing_comments.to_csv(f'ds_thematic-analysis/ds_stress-reducing-filtered_250comments_analysis.csv', index=False)
```

## data information

```python
df_stress_inducing = pd.read_csv('ds_thematic-analysis/ds_stress-inducing-filtered_comments.csv', low_memory=False, lineterminator='\n')
df_stress_reducing = pd.read_csv('ds_thematic-analysis/ds_stress-reducing-filtered_comments.csv', low_memory=False, lineterminator='\n')
```

```python
# Count the number of data points in each topic number
topic_counts = df_stress_inducing['top_weighted_topic'].value_counts()

# Create a bar chart
plt.bar(topic_counts.index.astype(str), topic_counts.values)

# Set labels and title
plt.xlabel('Topic Number')
plt.ylabel('Count')
plt.title('Count of comments that contain stress-inducing keywords per Topic Number')

# Rotate x-axis labels if needed
plt.xticks(rotation='vertical')

# Save the graph as a PNG image
plt.tight_layout()
plt.savefig('ds_thematic-analysis/stress-inducing-comments-count.png')

# Display the chart
plt.show()
```

```python
# Count the number of data points in each topic number
topic_counts_reducing = df_stress_reducing['top_weighted_topic'].value_counts()

# Create a bar chart
plt.bar(topic_counts_reducing.index.astype(str), topic_counts_reducing.values)

# Set labels and title
plt.xlabel('Topic Number')
plt.ylabel('Count')
plt.title('Count of comments that contain stress-reducing keywords per Topic Number')

# Rotate x-axis labels if needed
plt.xticks(rotation='vertical')

# Save the graph as a PNG image
plt.tight_layout()
plt.savefig('ds_thematic-analysis/stress-reducing-comments-count.png')

# Display the chart
plt.show()
```
